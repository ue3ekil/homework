```python
import pandas as pd
```

```python
movies = pd.read_csv('movies_stats.csv')
```

```python
years = list(range(1950, 2011))
```

```python
def film_production(title):
    title_str = str(title)
    for x in years:
        if str(x) in title_str:
            return x
    else:
        return 1900
```

```python
movies['year'] = movies['title'].apply(film_production)
```

```python
avg_rating = movies.groupby('year')['rating'].mean().sort_values(ascending=False).reset_index()
```
