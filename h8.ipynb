```python
import requests
```

```python
class CurrencyMax:
    def __init__(self):
        self.url = "https://www.cbr-xml-daily.ru/daily_json.js"
    
    def get_max(self):
        response = requests.get(self.url)
        data = response.json()
        
        max_currency_name = None
        max_rate = -1
        
        for currency_code, currency_data in data['Valute'].items():
            current_rate = currency_data['Value']
            if current_rate > max_rate:
                max_rate = current_rate
                max_currency_name = currency_data['Name']
                
        return max_currency_name
```

```python
finder = CurrencyMax()
result = finder.get_max()
print(result)
```
