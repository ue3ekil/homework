```python
import pandas as pd
```

```python
ratings = pd.read_csv('ratings.csv')
```

```python
movies = pd.read_csv('movies.csv')
```

```python
merged = ratings.merge(movies, on='movieId')
```

```python
counts = merged[merged['rating'] == 5.0] ['title'].value_counts()
```

```python
top_title = counts.index[0]
```

```python
print(top_title)
```
